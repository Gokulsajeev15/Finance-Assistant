```python
# Import libraries
# from google.colab import userdata
import openai
import os,time
from dotenv import load_dotenv
# Save your API key (secretly)
# openai.api_key = userdata.get('open_api_key')

load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
assistant = openai.beta.assistants.create(
    name="SmartInvest Assistant",
    instructions="You are a helpful financial advisor. Answer questions clearly and with relevant data.",
    model="gpt-4o-mini"
)

thread = openai.beta.threads.create()
print(f"Thread ID: {thread.id}")
```

```python
message = openai.beta.threads.messages.create(
    thread_id=thread.id,
    role="user",
    content="What is the gross margin % of Apple?"
)

# Ask the assistant to respond to the thread
run = openai.beta.threads.runs.create(
    thread_id=thread.id,
    assistant_id=assistant.id
)



while True:
    run_status = openai.beta.threads.runs.retrieve(
        thread_id=thread.id,
        run_id=run.id
    )
    if run_status.status == "completed":
        break
    time.sleep(1)

messages = openai.beta.threads.messages.list(thread_id=thread.id)

for msg in messages.data:
    if msg.role == "assistant":
        print("Assistant:", msg.content[0].text.value)
```
